```python
#Import phase
from subprocess import call #This will be used in order to make calls to our exes
import os
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
if os.path.isfile('../data/initialConditions.ic'):
    call(["../particlesToCSV.x", "../data/initialConditions.ic"])
    points = pd.read_csv("../data/initialConditions.ic.csv", sep=";")

    #fig = plt.figure()
    ax = plt.axes(projection="3d")
    points_x = points['x']
    points_y = points['y']
    points_z = points['z']
    ax.scatter3D(points_x, points_y, points_z, c=points_z, cmap='hsv');
    plt.show()
else:
    print("initialConditions.ic.csv not present")
```
